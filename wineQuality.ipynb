```python
red_wine_path = "/content/drive/MyDrive/datasets/wine+quality/winequality-red.csv"
white_wine_path = "/content/drive/MyDrive/datasets/wine+quality/winequality-white.csv"
```

```python
import pandas as pd

# Load datasets
red_wine = pd.read_csv(red_wine_path, sep=";")   # semicolon separator
white_wine = pd.read_csv(white_wine_path, sep=";")

# Show basic info
print("Red wine dataset:")
print(red_wine.head())

print("\nWhite wine dataset:")
print(white_wine.head())
```

```python
# Shape of datasets
print("Red wine shape:", red_wine.shape)
print("White wine shape:", white_wine.shape)

# Summary statistics
print(red_wine.describe())
print(white_wine.describe())

# Check for missing values
print(red_wine.isnull().sum())
print(white_wine.isnull().sum())
```

```python
red_wine["type"] = "red"
white_wine["type"] = "white"

# Combine into one dataset
wine = pd.concat([red_wine, white_wine], axis=0)

wine.head()
```

```python
from google.colab import sheets
sheet = sheets.InteractiveSheet(df=wine)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of wine quality by type
plt.figure(figsize=(6,4))
sns.countplot(x="quality", hue="type", data=wine)
plt.title("Wine Quality Distribution")
plt.show()
```
